```python
%pylab 
%matplotlib inline
import scipy
import seaborn as sns
sns.set_style('ticks')
```

```python
from scipy.stats import norm
figure(figsize=(5, 2.5))
x = linspace(-15, 15, 2000)
plot(x, norm(-5, 2.5).pdf(x))
plot(x, norm(5, 2.5).pdf(x))
axvline(0, color='k')
text(-0.5, .2, 'Choice', rotation=90)
axvline(-2.5, color='k')
text(-3, .229, 'Confidence', rotation=90)
axvline(2.5, color='k')
text(2.0, .229, 'Confidence', rotation=90)
yticks([])
sns.despine(left=True)
xlabel('Evidence')
savefig('/Users/nwilming/Desktop/f1.pdf', bbox_inches='tight')
```

```python
from scipy.stats import norm
figure(figsize=(5, 2.5))
x = linspace(-10, 10, 2000)
plot(x, norm(0, 3.5).cdf(x))
plot([-10, 10], [.5, .5], color='k', alpha=0.5)
text(10, 0.49, 'Choice')
plot([-10, 10], [.25, .25], color='k', alpha=0.5)
text(10, .229, 'Confidence')
plot([-10, 10], [.75, .75], color='k', alpha=0.5)
text(10, .729, 'Confidence')
yticks([0, .5, 1])
ylim([-.05, 1.05])
xlim([-10.5, 10.5])
sns.despine(trim=True)
xlabel('Sensory input')
ylabel('P(Yes | Data)')
savefig('/Users/nwilming/Desktop/f1.pdf', bbox_inches='tight')
```
